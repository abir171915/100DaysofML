### Step01: Data Pre-processing

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Social_Network_Ads.csv')
```

```python
X = df.iloc[:,[2,3]].values
y = df.iloc[:,4].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.2, random_state= 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

### Step02: Logistic Regression Model

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier.fit(X_train, y_train)
```

### Step03: Prediction

```python
y_pred = classifier.predict(X_test)
```

### Step04: Evaluating the Prediction

```python
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(y_test, y_pred)
```

```python
from matplotlib.colors import ListedColormap
def plot_decision_boundary(model, X, y, title="Decision Boundary"):
    # X must be (n_samples, 2)
    x1_min, x1_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    x2_min, x2_max = X[:, 1].min() - 1, X[:, 1].max() + 1

    xx1, xx2 = np.meshgrid(
        np.arange(x1_min, x1_max, 0.01),
        np.arange(x2_min, x2_max, 0.01)
    )

    grid = np.c_[xx1.ravel(), xx2.ravel()]
    Z = model.predict(grid).reshape(xx1.shape)

    plt.figure()
    plt.contourf(xx1, xx2, Z, alpha=0.25, cmap=ListedColormap(("red", "green")))
    plt.xlim(xx1.min(), xx1.max())
    plt.ylim(xx2.min(), xx2.max())

    for cls, color in zip(np.unique(y), ("red", "green")):
        plt.scatter(X[y == cls, 0], X[y == cls, 1], c=color, label=f"Class {cls}", s=20)

    plt.title(title)
    plt.xlabel("Feature 1")
    plt.ylabel("Feature 2")
    plt.legend()
    plt.show()
```

```python

plot_decision_boundary(classifier, X_test, y_test)
```
