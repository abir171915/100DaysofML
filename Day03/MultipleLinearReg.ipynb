### Step 01: Data Preprocessing

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("50_Startups.csv")
df
```

```python
X = df.iloc[:,:-1].values
y = df.iloc[:,4].values
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

ct = ColumnTransformer(
    transformers= [("encoder", OneHotEncoder(drop= "first"), [3])],
    remainder= "passthrough"
)

x = ct.fit_transform(X)
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size= 0.2, random_state= 0)
```

### Step 02: Fitting Multiple Linear Regression to the Training Set

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(X_train, y_train)
```

### Step 03: Predicting the Test set results 

```python
y_pred = reg.predict(X_test)
```

### Step 04 : plotting the result

```python
import matplotlib.pyplot as plt
plt.scatter(y_train, y_pred)
plt.xlabel("Actual y")
plt.ylabel("Predicted y")
plt.title("Actual vs Predicted (Train)")
plt.show()
```
